# 深圳6月aqi走势折线图

```python
sz=aqi[aqi['city'] == '深圳']
last_month=sz.sort_values('date').tail(34)
last_month=last_month[last_month['date']<'2018-07-01']
plt.figure(figsize=(16,8),dpi=80)
x=[i for i in range(1,31)]
plt.plot(x,last_month['aqi'].values)
plt.plot([10.98], [142.3], "o")
plt.plot([24], [28], "o")
y_ticks=range(500)
x_ticks=range(1,31)
plt.yticks(y_ticks[::30])
plt.ylabel('AQI数值',fontsize=15)
plt.xlabel('日期',fontsize=15)
plt.xticks(x_ticks)
plt.title('6月份深圳AQI指数走势',fontsize=18,color='black')
# 11 148
aqi_max=last_month['aqi'].max()
# 24 28
aqi_min=last_month['aqi'].min()
# last_month[last_month['aqi'] == 28]
plt.grid(linestyle='--',alpha=0.5)
plt.text(11,148,148,horizontalalignment='center',fontsize=15)
plt.text(24,25,28,horizontalalignment='left',verticalalignment='top',fontsize=15)
plt.savefig('6月份深圳AQI指数走势.png')
plt.show()
```

# 深圳 广州6月aqi数值柱形表,100基准

```python
gz=aqi[aqi['city'] == '广州']
gz=gz.sort_values('date')
gz_last_month=gz.tail(34)
gz_last_month=gz_last_month[gz_last_month['date']<'2018-07-01']
# gz_last_month
```

```python
# 数据准备,aqi以100为基准分组两组
great=last_month[last_month['aqi']>=100]
less = last_month[last_month['aqi']<100]
gz_great=gz_last_month[gz_last_month['aqi']>=100]
gz_less=gz_last_month[gz_last_month['aqi']<100]
plt.figure(figsize=(12,8),dpi=80)
plt.bar([0.9],[len(great['aqi'].values)],width=0.2,color='orange',label='轻度污染')
plt.bar([1.1],[len(less['aqi'].values)],width=0.2,color='lightblue',label='优良')
plt.bar([1.9],[len(gz_great['aqi'].values)],width=0.2,color='orange')
plt.bar([2.1],[len(gz_less['aqi'].values)],width=0.2,color='lightblue')
plt.legend()
plt.grid(linestyle='--',alpha=0.1)
# x_ticks=[i for i in range(1,4)]
plt.xticks(np.arange(4),['','深圳','广州',''])
plt.title('广深空气质量对比')
plt.savefig('广深空气质量对比.png')
plt.show()
```
